Task - 1:
Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.

```python
def convert_units(category, option, value):
    """
    Convert units based on category and option selected.

    Parameters:
    category (int): Type of conversion (1=Length, 2=Weight, 3=Volume)
    option (int): Specific conversion option
    value (float): Value to convert

    Returns:
    float: Converted result
    """

    if category == 1:  # Length
        if option == 1:
            return value * 3.28084   # meters to feet
        elif option == 2:
            return value / 3.28084   # feet to meters

    elif category == 2:  # Weight
        if option == 1:
            return value * 2.20462   # kg to lbs
        elif option == 2:
            return value / 2.20462   # lbs to kg

    elif category == 3:  # Volume
        if option == 1:
            return value * 0.264172  # liters to gallons
        elif option == 2:
            return value / 0.264172  # gallons to liters

    return None


try:
    print("Select Conversion Type")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    category = int(input("Enter your choice (1-3): "))

    if category == 1:
        print("\nLength Conversion")
        print("1. Meters (m) to Feet (ft)")
        print("2. Feet (ft) to Meters (m)")
        option = int(input("Choose option: "))
        value = float(input("Enter value: "))

        result = convert_units(category, option, value)

        if option == 1:
            print(value, "m =", result, "ft")
        elif option == 2:
            print(value, "ft =", result, "m")

    elif category == 2:
        print("\nWeight Conversion")
        print("1. Kilograms (kg) to Pounds (lbs)")
        print("2. Pounds (lbs) to Kilograms (kg)")
        option = int(input("Choose option: "))
        value = float(input("Enter value: "))

        result = convert_units(category, option, value)

        if option == 1:
            print(value, "kg =", result, "lbs")
        elif option == 2:
            print(value, "lbs =", result, "kg")

    elif category == 3:
        print("\nVolume Conversion")
        print("1. Liters (L) to Gallons (gal)")
        print("2. Gallons (gal) to Liters (L)")
        option = int(input("Choose option: "))
        value = float(input("Enter value: "))

        result = convert_units(category, option, value)

        if option == 1:
            print(value, "L =", result, "gal")
        elif option == 2:
            print(value, "gal =", result, "L")

    else:
        print("Invalid category selected.")

except ValueError:
    print("Invalid input! Please enter numeric values only.")
```

Task - 2:
Create a Python program that performs various mathematical operations on a list of numbers.
• The Program should:
1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum
of the numbers).
2. Ask the user to input a list of numbers (separated by spaces).

3. Perform the selected operation and display the result.
4. Handle potential errors, such as invalid input or empty lists

```python
def find_sum(numbers):
    """
    Calculate the sum of a list of numbers.

    Parameters:
    numbers (list): List of numeric values.

    Returns:
    float: Sum of the numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculate the average of a list of numbers.

    Parameters:
    numbers (list): List of numeric values.

    Returns:
    float: Average value.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Find the maximum number in a list.

    Parameters:
    numbers (list): List of numeric values.

    Returns:
    float: Maximum value.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Find the minimum number in a list.

    Parameters:
    numbers (list): List of numeric values.

    Returns:
    float: Minimum value.
    """
    return min(numbers)


try:
    print("Choose the operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = int(input("Enter your choice (1-4): "))

    numbers_input = input("Enter numbers separated by spaces: ")

    numbers = [float(num) for num in numbers_input.split()]

    if len(numbers) == 0:
        print("Error: The list is empty.")
    else:
        if choice == 1:
            print("Sum =", find_sum(numbers))
        elif choice == 2:
            print("Average =", find_average(numbers))
        elif choice == 3:
            print("Maximum =", find_maximum(numbers))
        elif choice == 4:
            print("Minimum =", find_minimum(numbers))
        else:
            print("Invalid choice.")

except ValueError:
    print("Error: Please enter valid numbers only.")
```

**4.3 Exercise on Nested List:**

**1. Flatten a Nested List:**
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

```python
def flatten(lst):
    """
    Flatten a nested list into a single list.

    Parameters:
    lst (list): A nested list containing sublists.

    Returns:
    list: A flattened list containing all elements.
    """
    flat_list = []

    for sublist in lst:
        for item in sublist:
            flat_list.append(item)

    return flat_list


# Example
numbers = [[1, 2], [3, 4], [5]]
result = flatten(numbers)

print("Original nested list:", numbers)
print("Flattened list:", result)
```

**2. Accessing Nested List Elements:**
Write a Python function that extracts a specific element from a nested list given its indices.

```python
def access_nested_element(lst, indices):
    """
    Access a specific element from a nested list using given indices.

    Parameters:
    lst (list): The nested list.
    indices (list): A list of indices showing the position of the element.

    Returns:
    element: The element at the specified position.
    """
    element = lst

    for index in indices:
        element = element[index]

    return element


# Example
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]

result = access_nested_element(lst, indices)

print("Nested List:", lst)
print("Indices to access:", indices)
print("Element at position", indices, "is:", result)
```

**3. Sum of All Elements in a Nested List:**
Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

```python
def sum_nested(lst):
    """
    Calculate the sum of all numbers in a nested list.

    Parameters:
    lst (list): A nested list that may contain numbers or other lists.

    Returns:
    int/float: Sum of all numbers in the nested list.
    """
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)   # recursive call for sublist
        else:
            total += item

    return total


# Example
numbers = [[1, 2], [3, [4, 5]], 6]

result = sum_nested(numbers)

print("Nested List:", numbers)
print("Sum of all elements:", result)
```

**4. Remove Specific Element from a Nested List:**
Write a Python function that removes all occurrences of a specific element from a nested list.

```python
def remove_element(lst, elem):
    """
    Remove all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): A nested list that may contain other lists.
    elem: The element to remove.

    Returns:
    list: Modified list with the element removed.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))  # process sublist
        elif item != elem:
            result.append(item)

    return result


# Example
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2

result = remove_element(lst, elem)

print("Original Nested List:", lst)
print("Element to remove:", elem)
print("Modified List:", result)
```

**5. Find the Maximum Element in a Nested List:**
Write a Python function that finds the maximum element in a nested list (regardless of depth).

```python
def find_max(lst):
    """
    Find the maximum element in a nested list.

    Parameters:
    lst (list): A nested list that may contain numbers or other lists.

    Returns:
    int/float: The maximum element in the nested list.
    """
    maximum = None

    for item in lst:
        if isinstance(item, list):
            value = find_max(item)  # recursive call
        else:
            value = item

        if maximum is None or value > maximum:
            maximum = value

    return maximum


# Example
numbers = [[1, 2], [3, [4, 5]], 6]

result = find_max(numbers)

print("Nested List:", numbers)
print("Maximum element in the nested list:", result)
```

**6. Count Occurrences of an Element in a Nested List:**
Write a Python function that counts how many times a specific element appears in a nested list.

```python
def count_occurrences(lst, elem):
    """
    Count how many times a specific element appears in a nested list.

    Parameters:
    lst (list): A nested list that may contain numbers or other lists.
    elem: The element to count.

    Returns:
    int: Number of times the element appears.
    """
    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)  # recursive call
        elif item == elem:
            count += 1

    return count


# Example
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2

result = count_occurrences(lst, elem)

print("Nested List:", lst)
print("Element to count:", elem)
print("Number of occurrences:", result)
```

**7. Flatten a List of Lists of Lists:**
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

```python
def deep_flatten(lst):
    """
    Flatten a deeply nested list into a single list.

    Parameters:
    lst (list): A list that may contain other nested lists.

    Returns:
    list: A single flattened list containing all elements.
    """
    flat_list = []

    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # recursive call
        else:
            flat_list.append(item)

    return flat_list


# Example
lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

result = deep_flatten(lst)

print("Original Nested List:", lst)
print("Flattened List:", result)
```

**8. Nested List Average:**
Write a Python function that calculates the average of all elements in a nested list.

```python
def average_nested(lst):
    """
    Calculate the average of all elements in a nested list.

    Parameters:
    lst (list): A nested list containing numbers or other lists.

    Returns:
    float: The average of all elements.
    """
    total = 0
    count = 0

    for item in lst:
        if isinstance(item, list):
            sub_total, sub_count = average_nested(item)
            total += sub_total
            count += sub_count
        else:
            total += item
            count += 1

    return total, count


# Example
lst = [[1, 2], [3, 4], [5, 6]]

total, count = average_nested(lst)
average = total / count

print("Nested List:", lst)
print("Average of all elements:", average)
```

```python
import numpy as np

A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

print("Matrix A:\n", A)
print("\n Matrix B:\n", B)
```

```python
# Inverse of A
A_inv = np.linalg.inv(A)

# Multiply A with its inverse
result1 = np.dot(A, A_inv)

# Identity matrix for comparison
I = np.eye(2)

print("A * A_inv:\n", result1)
print("\n Identity Matrix I:\n", I)

# Check if result is close to identity
print("\n Is A * A_inv approximately equal to I?", np.allclose(result1, I))
```
